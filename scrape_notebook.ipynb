```python
# Given an article URL, extract the page content

from rich import print
import requests

url = "https://www.ft.com/semiconductors?page=2"

article_url = "https://www.ft.com/content/3dd07583-21f7-42ec-be8f-78c58279ecc4"

archive_prefix = "https://archive.md/"

req = requests.get(article_url)

req2 = requests.get(archive_prefix + article_url)
print(req)
```

```python
from bs4 import BeautifulSoup
from bs4.element import Comment
from typing import Any

def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def content_to_text(content: bytes | Any) -> str:
    bs = BeautifulSoup(content, 'html.parser')
    texts = bs.find_all(string=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)

print(content_to_text(req2.content))
```
